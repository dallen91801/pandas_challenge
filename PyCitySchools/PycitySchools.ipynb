```python
import pandas as pd
from pathlib import Path
```

```python
school_data_to_load = Path("C:/Users/board/pandas_challenge/PyCitySchools/Resources/schools_complete.csv")
student_data_to_load = Path("C:/Users/board/pandas_challenge/PyCitySchools/Resources/students_complete.csv")
```

```python
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
   
```

```python
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name"])
```

```python
total_schools = school_data_complete['school_name'].nunique()
```

```python
total_students = school_data_complete['Student ID'].nunique()
```

```python
total_budget = school_data['budget'].sum()
```

```python
average_math_score = school_data_complete['math_score'].mean()
```

```python
average_reading_score = school_data_complete['reading_score'].mean()
```

```python
passing_math = school_data_complete[school_data_complete['math_score'] >= 70].count()['Student ID']
percent_passing_math = (passing_math / total_students) * 100
```

```python
passing_reading = school_data_complete[school_data_complete['reading_score'] >= 70].count()['Student ID']
percent_passing_reading = (passing_reading / total_students) * 100
```

```python
overall_passing = school_data_complete[
    (school_data_complete['math_score'] >= 70) & 
    (school_data_complete['reading_score'] >= 70)
].count()['Student ID']
percent_overall_passing = (overall_passing / total_students) * 100
```

```python
district_summary = pd.DataFrame({
    "Total Schools": [total_schools],
    "Total Students": [total_students],
    "Total Budget": [total_budget],
    "Average Math Score": [average_math_score],
    "Average Reading Score": [average_reading_score],
    "% Passing Math": [percent_passing_math],
    "% Passing Reading": [percent_passing_reading],
    "% Overall Passing": [percent_overall_passing]})
```

```python
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)
```

```python
# Creating the spending_summary DataFrame
spending_summary = pd.DataFrame({
    "Average Math Score": spending_math_scores,
    "Average Reading Score": spending_reading_scores,
    "% Passing Math": spending_passing_math,
    "% Passing Reading": spending_passing_reading,
    "% Overall Passing": overall_passing_spending
})

# Display the spending_summary DataFrame
spending_summary
```
